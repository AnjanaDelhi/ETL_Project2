```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as numpy
import psycopg2
```

```python
billboard_csv = "Billboard.csv"
billboard_df = pd.read_csv(billboard_csv)
billboard_df.head()
```

```python
billboard2_df = billboard_df.rename(columns={"Song": "Song/Nominee"})
billboard2_df.head()
```

```python
# removed Unnamed: 0 column
billboard_updated_df = billboard2_df.drop("Unnamed: 0", axis = "columns")
billboard_updated_df.head()
```

```python
grammy_csv = "the_grammy_awards.csv"
grammy_df = pd.read_csv(grammy_csv)
grammy_df.head(2)
```

```python
grammy2_df = grammy_df.rename(columns={"artist": "Artist", "nominee": "Song/Nominee"})
grammy2_df.head()
```

```python
# removed published_at, updated_at, workers, and img columns
grammy_update_df = grammy2_df.drop(["published_at", "title", "updated_at", "workers", "img"], axis = "columns")
grammy_update_df.head()
```

```python
merge_df = billboard_updated_df.merge(grammy_update_df, on= ["Song/Nominee", "Artist"])
merge_df.head(15)
```

```python
df_clean = merge_df.drop_duplicates()
df_clean.head(15)
```

```python
df_clean = merge_df.drop_duplicates( "Song/Nominee" , keep='first')
df_clean.head(15)
```

```python
weeks_artist = df_clean.groupby("Artist")["Weeks On #1", "Weeks On Chart"].sum().reset_index()

weeks_artist.head()
```

```python
#Rename columns to better names for transferability to SQL database
weeks_artist2 = weeks_artist.rename(columns={"Artist": "artist", "Weeks On #1": "weeks_at1", "Weeks On Chart": "weeks"})
weeks_artist2.head()
```

```python
weeks_song = df_clean.groupby("Song/Nominee")["Weeks On #1", "Weeks On Chart"].sum().reset_index()
weeks_song.head()
```

```python
#Rename columns to better names for transferability to SQL database
weeks_song2 = weeks_song.rename(columns={"Song/Nominee": "song_nominee", "Weeks On #1": "weeks_at1", "Weeks On Chart": "weeks"})
weeks_song2.head()
```

```python
awards_weeks = df_clean.groupby(["Song/Nominee", "category"])["Weeks On #1", "Weeks On Chart"].sum().reset_index()
awards_weeks.head()
```

```python
awards_weeks2 = awards_weeks.rename(columns={"Song/Nominee": "song_nominee" ,  "Weeks On #1": "weeks_at1", "Weeks On Chart": "weeks"})
awards_weeks2.head()
```
